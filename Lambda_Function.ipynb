### Normal Function

```python
def square(x):
    return x**2
square(5)
```

### lambda Function

```python
square=lambda x:x**2
square(5)
```

#### map function

```python
number=[1,2,3,4]
square=list(map(lambda x:x**2,number))
square
```

#### filter function

```python
numbers=[1,2,3,4,5,6,7,8]
evens=filter(lambda x:x%2==0,numbers)
print(list(evens))
```

```python
numbers=[1,2,3,4,5,6,7,8]
evens=map(lambda x:x%2==0,numbers)
print(list(evens))
```

```python
number=[1,2,3,4]
square=list(filter(lambda x:x**2,number))
square
```

#### reduce function

```python
from functools import reduce
numbers=[1,2,3,4]
product=reduce(lambda x,y:x*y,numbers)
product
```

#### lambda with multiple argument

```python
add=lambda x,y:x+y
add(3,4)
```

```python
maximum_value=lambda x,y:x if x>y else y
maximum_value(10,5)
```

#### Sorting function in lambda

```python
students=[("alice",85),("bob",78),("charlie",92)]
students.sort(key=lambda students:students[1],reverse=True)
print(students)
```

```python
students=[("alice",85),("bob",78),("charlie",92)]
students.sort(key=lambda students:students[1])
print(students)
```

## Decorators

```python
def morning_decorator(func):
    def wrapper():
        print("Good Morning")
        func()
        print("Thanks")
    return wrapper
```

```python
def greet():
    print("Hello")
greet()
```

### Manual Way

```python
greet=morning_decorator(greet)
greet()
```

### Syntax Method

```python
@morning_decorator
def greet():
    print("Hello")
greet()
```
